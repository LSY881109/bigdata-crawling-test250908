```python
#bs4 (BeautifulSoup)
#   HTML XML 
#      .
#      
#          .

##  
#    (tag)    .
# BeautifulSoup      
#    (Parse Tree)  .
#              .
#
#  ,
#   '' ,
# BeautifulSoup  ()  ,
#    (, )    
#  '' .

# 
# BeautifulSoup  
#       .
#
# 1. BeautifulSoup  
#  HTML   soup   .
from bs4 import BeautifulSoup

#   HTML 
html_doc = "<html><head><title>The Dormouse's story</title></head><body>...</body></html>"

# BeautifulSoup   (HTML , ' ')
soup = BeautifulSoup(html_doc, 'html.parser')


# 2. (Tag) : find() find_all()
# soup.find(''):        .
# soup.find_all(''):      (List)  .
#    <p>   
p_tag = soup.find('p')
#  <li>  
li_tags = soup.find_all('li')

# 3  (Text) : .text
#   .text  
#        .
# <title>   
title_text = soup.find('title').text
print(title_text)  # : The Dormouse's story

# 4.  (Attribute)  
#  (:   href,   src)
#   []  .
# <a>  ,   href   
a_tag = soup.find('a')
link_url = a_tag['href']
print(link_url)

#  5 CSS  : select() select_one()
# CSS    select()   .
#
# soup.select_one(''): CSS      .
#
# soup.select(''): CSS      .
```
